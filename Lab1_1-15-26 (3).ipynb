# Reflection questions 7
### Was the object mutable or immutable?
- **Mutable**.  
  Lists in Python can be changed in place.

### Did Python mutate the object or rebind the name?
- reset_wrong: Python rebound the name lst.  
  The assignment lst = [] makes lst point to a new list, leaving the original list unchanged.

- reset_right Python mutated the object.  
  The method lst.clear() empties the existing list in place.


```python
# example 7
def reset_wrong(lst):
    lst = []

def reset_right(lst):
    lst.clear()

data = [1, 2, 3]
reset_wrong(data)
print(data)

reset_right(data)
print(data)
```

- Neither.
